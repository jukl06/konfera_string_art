```python
import cv2
import numpy as np
from make_stringart import make_stringart
from matplotlib import pyplot as plt
```

```python
cap = cv2.VideoCapture("super_trim_rick.mp4")

lpnb = 0
res = (640, 480)

fourcc = cv2.VideoWriter_fourcc(*"XVID")
# out = cv2.VideoWriter("line_rick.avi", fourcc, 20, res)
out = cv2.VideoWriter("line_rick.mp4", 0, 20, res)
```

```python
while cap.isOpened():
    print(lpnb)
    lpnb += 1
    if lpnb < 60:
        continue
    ret, frame = cap.read()
    if not ret:
        break

    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    stred = (frame.shape[0] // 2, frame.shape[1] // 2)
    velikost = min(frame.shape[0], frame.shape[1]) // 2
    frame = frame[
        stred[0] - velikost : stred[0] + velikost,
        stred[1] - velikost : stred[1] + velikost,
    ]

    frame = cv2.resize(frame, res)
    frame = cv2.resize(frame, (250, 250))

    out_img = make_stringart(
        cv2.cvtColor(frame, cv2.COLOR_GRAY2BGR),
        32,
        1,
        resolution=(frame.shape[0], frame.shape[1]),
    )
    out_img = cv2.resize(out_img, (velikost * 2, velikost * 2))

    out_img_padded = np.full((res[1], res[0]), 255, dtype=np.uint8)
    h, w = out_img.shape
    y_offset = (res[1] - h) // 2
    x_offset = (res[0] - w) // 2
    out_img_padded[y_offset : y_offset + h, x_offset : x_offset + w] = out_img
    out_img = out_img_padded

    # plt.imshow(out_img)
    cv2.imwrite(f"imgs/img_{lpnb:04d}.png", out_img)
    # out.write(out_img)
    # break
```

```python
cap.release()
out.release()
```
